```python
import os
import shutil
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoAlertPresentException
import time

chrome_options = Options()
chrome_options.add_argument('--ignore-certificate-errors')

# Inicializar el navegador
driver = webdriver.Chrome(options=chrome_options)



# Abrir la página de inicio de sesión
driver.get('https://pagina-a-la-que-se-requiere-ingresar-login.jsp')

# Ingresar las credenciales y enviar el formulario
username = driver.find_element(By.ID, 'id')
password = driver.find_element(By.ID, 'passwd')

username.send_keys('usuario')
password.send_keys('Contraseña')

login_button = driver.find_element(By.NAME, 'SUBMIT_BUTTON')
login_button.click()

elementReporte = driver.find_element(By.CLASS_NAME, 'recentReportImg')
elementReporte.click()



def descargarBaseDatos (nomBD):
    
        ruta_carpeta_descargas = 'C:\\Users\\Downloads\\'  # Ruta de la carpeta de descargas
        cantidad_archivos_anterior = len(os.listdir(ruta_carpeta_descargas))  # Número inicial de archivos en la carpeta descargas

        elementoSelect = driver.find_element(By.ID, 'xdo:_paramstarget_name_div')
        elementoSelect.click()

        elementoBuscar = driver.find_element(By.CLASS_NAME, 'dialogFloatRight')
        elementoBuscar.click()

        nombreBD = driver.find_element(By.ID, 'xdo:xdo:_paramstarget_name_div_input_searchDialog_input')
        nombreBD.send_keys(nomBD)

        botonBuscar= driver.find_element(By.ID, 'xdo:xdo:_paramstarget_name_div_input_searchDialog_button')
        botonBuscar.click()

        time.sleep(2)
        seleccionarBD = driver.find_element(By.ID, 'xdo:xdo:_paramstarget_name_div_input_searchDialog_searchResults')
        seleccionarBD.click()

        botonAceptarA = driver.find_element(By.CLASS_NAME, 'dialog_footer-content')
        botonesAceptarB = botonAceptarA.find_elements(By.CLASS_NAME, 'button')
        botonAceptarB = botonesAceptarB[0]
        botonAceptarB.click()
        time.sleep(2)

        botonAplicar = driver.find_element(By.ID, 'reportViewApply')
        botonAplicar.click()
        time.sleep(1)

        opcionesExport = driver.find_element(By.ID, 'xdo:viewFormatIcon')
        opcionesExport.click()
        seleccionarExcel = driver.find_element(By.ID, '_xdoFMenu5')
        seleccionarExcel2 = seleccionarExcel.find_elements(By.TAG_NAME, 'a')
        quinto_a = seleccionarExcel2[4]
        quinto_a.click()
            
        time.sleep(1)

        # Esperar hasta que se agregue un nuevo archivo en la carpeta de descargas
        espera = WebDriverWait(driver, 8)  # Espera un máximo de 8 segundos
  
        espera.until(lambda driver: len(os.listdir(ruta_carpeta_descargas)) > cantidad_archivos_anterior)
            
```

```python
def eliminarArchivos():
    ruta_destino = 'C:\\Users\\Documents\\Cruces Python - doble validacion\\Data\\BD\\' #carpeta a la que se quiere mover los archivos
    archivos_eliminar = os.listdir(ruta_destino)

    # Eliminar cada archivo dentro de la carpeta hasta que quede vacia
    for archivo in archivos_eliminar:
        ruta_archivo = os.path.join(ruta_destino, archivo)
        if os.path.isfile(ruta_archivo):
            os.remove(ruta_archivo)


def moverArchivo(nomArchivo):

    carpeta_descargas = 'C:\\Users\\Downloads\\'
    ruta_destino = 'C:\\Users\\Documents\\Cruces Python - doble validacion\\Data\\BD\\'#carpeta a la que se quiere mover los archivos
    nomArchivo
    # Obtener la lista de archivos en la carpeta de descargas

    archivos_descargas = os.listdir(carpeta_descargas)

    # Ordenar los archivos por fecha de modificación de forma descendente
    archivos_descargas.sort(key=lambda x: os.path.getmtime(os.path.join(carpeta_descargas, x)), reverse=True)

    # Obtener el último archivo descargado
    ultimo_archivo = archivos_descargas[0]

    # Crear la ruta completa del último archivo descargado
    ruta_ultimo_archivo = os.path.join(carpeta_descargas, ultimo_archivo)

    # Mover el último archivo a la carpeta de destino
    shutil.move(ruta_ultimo_archivo, ruta_destino)
    
    # Obtener la lista de archivos en la carpeta destino
    archivos_bd = os.listdir(ruta_destino)

    # Ordenar los archivos por fecha de modificación de forma descendente
    archivos_bd.sort(key=lambda x: os.path.getmtime(os.path.join(ruta_destino, x)), reverse=True)

    # Obtener el último archivo que se movió a la carpeta
    ultimo_archivo_bd = archivos_bd[0]

    # Cambiar el nombre del archivo en la nueva ubicación

    ruta_nuevo_archivo = os.path.join(ruta_destino, nomArchivo)
    os.rename(os.path.join(ruta_destino, ultimo_archivo_bd), ruta_nuevo_archivo)


    driver.refresh()
    time.sleep(1)

```

```python
eliminarArchivos()

#Diccionario con el nombre del archivo que queremos descargar y el nombre que le queremos poner al archivo
bd={'REMEDY_REPPDB':'MYIT - REPPDB.xlsx', 'IDSPRO1_lnxracdbpr01':'CA - LNXRACDBPR01.xlsx', 'OIM1_comsun68':'OIM - OIM1_comsun68.xlsx', 'CALDMPROD_comibm17':'CALLIDUS - CALDMPROD.xlsx', 'TRUECPROD_comibm17':'CALLIDUS - TRUECPROD.xlsx', 'LOMBAPROD_comibm17':'CALLIDUS - LOMBAPROD.xlsx', 'BUSOBPROD_comibm17':'CALLIDUS - BUSOBPROD.xlsx', 'CDBRACDB_lnxracdbpr_PDB_WBMSDBPR':'INTEC - WBMSDBPR.xlsx', 'INTEGPRD2_comsun70':'INTEGRADOR WEB - INTEGPRD 70.xlsx', 'INTEGPRD1_comsun71': 'INTEGRADOR WEB - INTEGPRD 71.xlsx', 'GPRS_comsun101':'MAGUARE  - GPRS.xlsx', 'prtpcol1_comibm108':'MAHINDRA - PRTPCOL1_comibm108.xlsx', 'prtpcol2_comibm107':'MAHINDRA - PRTPCOL2_comibm107.xlsx', 'ussdcol1_lnxdbussd01':'MAHINDRA - ussdcol1.xlsx', 'ussdcol2_lnxdbussd02':'MAHINDRA - ussdcol2.xlsx', 'DATAREP_comhp43':'MOTOR DE PAGOS - DATAREP.xlsx', 'ACTIVA_comhp06':'POLIEDRO - ACTIVA.xlsx', 'POLIEXP_comhp06':'POLIEDRO - POLIEXP.xlsx', 'cofacprd_colbtasga01':'SGA - COFACPRD.xlsx', 'CAJAS_comhp01':'SICACOM - CAJAS.xlsx', 'CTGINS1_exacc101vm01': 'MAXIMO - CTGINS1_exacc101vm01.xlsx', 'REMEDY_REMPDB':'MYIT - REMPDB.xlsx', 'REMEDY_PDB_SMARTIT':'MYIT - PDB_SMARTIT.xlsx', 'REMEDY_DWPPDB':'MYIT - DWPPDB.xlsx', 'REMEDY_PDB_DWP':'MYIT - PDB_DWP.xlsx'}

#recorremos el diccionario
for a, b in bd.items():
    while True:
        
        time.sleep(2)
        try:
            print(a)
            descargarBaseDatos(a)    #Se busca la base de datos que requerimos
            print(b)
            moverArchivo(b)             #luego de descargado el archivo, lo movemos y le cambiamos el nombre
        except:
            print('superó el tiempo')   #si se presenta un error al descargar el archivo, se refresca la pagina y se vuelve a intentar
            try:
                driver.refresh()
                WebDriverWait(driver, 5).until(EC.alert_is_present())   
                driver.switch_to.alert.accept()                         #si al refrescar la pagina se muestra una alerta, esta es aceptada
            except NoAlertPresentException:
                pass
        else:
            break

    
```
